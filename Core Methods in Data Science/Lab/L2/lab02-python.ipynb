## Q2 Explore deaths in Washington

### 1. Take the subset you created above and filter results for Washington state only. Store it into a new variable.

```python
waDf = newDf[newDf.Province_State == "Washington"]
waDf
```

### 2. Add a new variable, “growth”, to the WA data. It should be the growth in the cumulative number of deaths through the last year, between 10/6/22 and 3/15/2021.

```python
waDf["growth"] = waDf["10/6/22"] - waDf["3/15/21"]
waDf
```

### 3. You may notice a warning: “A value is trying to be set on a copy of a slice...”. Why do we get this warning? How can we get rid of it? Repeat the previous steps in a way that you do not get the warning!

The reason why we get this warning is because the command line above returns the view of dataframe that may cause problem when filtering the data later. To get rid of this message, we need to put ".copy()" at the end of the command line to create unique dataframe.

```python
waDf = newDf[newDf.Province_State == "Washington"].copy()
waDf["growth"] = waDf["10/6/22"] - waDf["3/15/21"]
waDf
```

### 4. Extract the newly created variable “growth” as series. Show that it is a series!
